```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan

# --- 1. Load Dataset ---
df = pd.read_csv("50_Startups.csv")  # <- change file name if needed
X = df[['Administration']]                    # Independent variable
y = df['Profit']                   # Dependent variable

# --- 2. Fit scikit-learn model ---
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
residuals = y - y_pred

# --- 3. Residual Plot ---
plt.scatter(y_pred, residuals, color='darkorange', edgecolor='k')
plt.axhline(y=0, color='gray', linestyle='--')
plt.xlabel("Predicted Salary")
plt.ylabel("Residuals")
plt.title("Residual Plot (Administration vs. profit)")
plt.show()
```
