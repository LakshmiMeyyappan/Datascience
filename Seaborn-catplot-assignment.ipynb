```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('exercise')
sb.catplot(x = "time", y = "pulse", hue = "kind", kind = 'violin', col = "diet", data = df);
plt.show()
```

## Summarisation

```python
# In the above graph, pulse rate is influenced by exercise duration, type, and diet, using violin plots.

# For no fat diet, 

# At 1 min, pulse rate for running spread is high compared to rest and walking.

# By 15 min, 

#Running leads to a significant spike in pulse from (100 - 150), with a wide distribution.
#Compared to rest and walking, running leads to a significant spike.

# By 30 mins,
#Running leads to a significant spike in pulse from (120 - 160), with a wide distribution.
#Compared to rest and walking, running leads to a significant spike.


#For Low fat diet,

# At 1 min, the distribution of walking is high, compared to rest and running.

# By 15 mins, 
# distribution of rest is high.
# pulse rate for running is very high(80-150), Compared to rest and walking.

# By 30 mins,
# pulse rate for walking is from (75-115), 
# pulse rate for running is starting from (85-130)

# Across both diet types (low fat and no fat), the general pattern is the same.

# The longer and more intense the exercise, the higher the pulse.
```
